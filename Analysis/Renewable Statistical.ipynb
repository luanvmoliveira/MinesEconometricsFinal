```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import seaborn as sns
```

```python
df_renewable = pd.read_csv("treated_data/renewable_treated.csv",sep = ";", header = 0)
df_renewable_month = pd.read_csv("treated_data/renewable_month_treated.csv",sep = ";",header=0)
```

### First Contact

```python
print(plt.hist(df_renewable_month['Value']))
```

```python
print(plt.hist(df_renewable['Value']))
```

```python
ren_desc = df_renewable['Value'].describe()
ren_month_desc = df_renewable_month.groupby('Month')['Value'].describe()
print(f'Description de la production de Energies Renouvelables: Big picture\n\n{ren_desc}\n')
print(sns.boxplot(data = df_renewable['Value']))
```

### Production over time

```python
sns.lineplot(x = 'Year', y = 'Value', data = df_renewable)
```

```python
#Evolution
sns.lineplot(x = 'YYYYMM', y = 'Value', data = df_renewable_month)
```

### Descriptive statistiques in-between months

```python
print(f'Distribuition des statistiques descriptives parmis les mois:\n\n{ren_month_desc[:12]}\n\n')
print(sns.boxplot(data=ren_month_desc[:12]))
```

```python
#Variation saisonnière
sns.lineplot(x='Month', y = 'Value', data = df_renewable_month.groupby('Month').mean())
```
