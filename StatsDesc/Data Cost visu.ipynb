```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_CO2CA_prices = pd.read_csv('treated_data/Cost/california_co2_prices.csv', sep = ';')
df_CO2RGGI_prices = pd.read_csv('treated_data/Cost/rggi_co2_prices.csv', sep = ';')
df_coal_prices = pd.read_csv('treated_data/Cost/coal_month_treated.csv', sep = ';')
df_gas_prices = pd.read_csv('treated_data/Cost/gaz_month_treated.csv', sep = ';')
```

```python
CO2CA_prices = df_CO2CA_prices.iloc[:, 4].to_numpy()
CO2CA_dates = df_CO2CA_prices['dates'].to_numpy()
CO2RGGI_prices = df_CO2RGGI_prices.iloc[:, 4].to_numpy()
CO2RGGI_dates = df_CO2RGGI_prices['dates'].to_numpy()
coal_prices = df_coal_prices['Value'].to_numpy()
coal_dates = df_coal_prices['YYYYMM'].to_numpy()
gas_prices = df_gas_prices['Value'].to_numpy()
gas_dates = df_gas_prices['YYYYMM'].to_numpy()
```

```python
plt.plot(CO2CA_dates, CO2CA_prices)
```

```python
plot = sns.lineplot(x = 'Year', y = 'CO2_prices', data = df_CO2CA_prices)
plot.set_ylabel('$/ton')
plot.set_xlabel('Time')
plot.set_title('Price of CO2 in California - monthly')
```

```python
plot = sns.lineplot(x = 'Year', y = 'CO2_prices', data = df_CO2RGGI_prices)
plot.set_ylabel('$/ton')
plot.set_xlabel('Time')
plot.set_title('Price of CO2 in RGGI - monthly')
```

```python
plot = sns.lineplot(x = 'Year', y = 'Value', data = df_coal_prices)
plot.set_ylabel('$/ton')
plot.set_xlabel('Time')
plot.set_title('Price of Coal in the US - monthly')
```

```python
plot = sns.lineplot(x = 'Year', y = 'Value', data = df_gas_prices)
plot.set_ylabel('$/ton')
plot.set_xlabel('Time')
plot.set_title('Price of Gas in the US - monthly')
```
